```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from datetime import datetime

#this is a python magic function to automatically show matplotlib plots
%matplotlib inline   
```

```python
june_trips = pd.read_pickle('../data/june_trips.pkl')
may_trips = pd.read_pickle('../data/may_trips.pkl')
july_trips = pd.read_pickle('../data/july_trips.pkl')
```

```python
tripcolnames = ['tripRoute', 'startDateTime_clean', 'endDateTime_clean', 'sumdID', 'startLatitude', 'startLongitude', 'endLatitude', 'endLongitude']
may_tripsdup = may_trips.duplicated(subset = tripcolnames, keep = False).sum()
may_tripsdup
```

```python
june_tripsdup = june_trips.duplicated(subset = tripcolnames, keep = False).sum()
june_tripsdup
```

```python
july_tripsdup = july_trips.duplicated(subset = tripcolnames, keep = False).sum()
july_tripsdup
```

```python
may_trips_dedup = may_trips.drop_duplicates(subset = ['tripRoute', 'startDateTime_clean', 'endDateTime_clean', 'sumdID', 'startLatitude', 'startLongitude', 'endLatitude', 'endLongitude'], keep = 'first')
may_trips_dedup.head()
```

```python
june_trips_dedup = june_trips.drop_duplicates(subset = ['tripRoute', 'startDateTime_clean', 'endDateTime_clean', 'sumdID', 'startLatitude', 'startLongitude', 'endLatitude', 'endLongitude'], keep = 'first')
june_trips_dedup.head()
```

```python
july_trips_dedup = july_trips.drop_duplicates(subset = ['tripRoute', 'startDateTime_clean', 'endDateTime_clean', 'sumdID', 'startLatitude', 'startLongitude', 'endLatitude', 'endLongitude'], keep = 'first')
july_trips_dedup.head()
```
